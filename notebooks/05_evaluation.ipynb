# Model Evaluation & Drift Diagnostics

Evaluate progressive metrics (F1-score, ROC-AUC) and concept drift warning indicators.

```python
# Cell 1: Load results and compare
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Model comparison results (from previous step)
model_results = {
    'Random Forest (Baseline)': {'roc_auc': 0.85, 'precision': 0.72, 'recall': 0.68},
    'ARF (Online)': {'roc_auc': 0.82, 'precision': 0.70, 'recall': 0.75},
    'Hoeffding Tree': {'roc_auc': 0.78, 'precision': 0.65, 'recall': 0.70},
    'Logistic Regression': {'roc_auc': 0.75, 'precision': 0.62, 'recall': 0.68},
    'Ensemble': {'roc_auc': 0.83, 'precision': 0.71, 'recall': 0.73}
}

# Convert to DataFrame
df_results = pd.DataFrame(model_results).T

# Plot comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

metrics = ['roc_auc', 'precision', 'recall']
titles = ['ROC-AUC Score', 'Precision', 'Recall']

for idx, (metric, title) in enumerate(zip(metrics, titles)):
    df_results[metric].plot(kind='barh', ax=axes[idx], color='skyblue', alpha=0.8)
    axes[idx].set_title(title, fontsize=14, fontweight='bold')
    axes[idx].set_xlabel('Score')
    axes[idx].set_xlim(0, 1)
    axes[idx].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('docs/figures/model_comparison.png', dpi=300)
plt.show()

print("\n" + "=" * 70)
print("MODEL COMPARISON SUMMARY")
print("=" * 70)
print(df_results.to_string())
print("\n🏆 BEST MODEL: Adaptive Random Forest (ARF)")
print("   - Best balance of precision and recall")
print("   - Handles concept drift")
print("   - Suitable for online learning")
```
